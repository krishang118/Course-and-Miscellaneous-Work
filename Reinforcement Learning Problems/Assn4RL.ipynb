```python
import numpy as np
import matplotlib.pyplot as plt
arrivalrates = [0.5, 0.6, 0.7, 0.8, 0.9]
queueslen = []
responsetimes = []
for i in arrivalrates:
    et = 1 / (1 - i)
    en = i / (1 - i)
    queueslen.append(en)
    responsetimes.append(et)
print("The values are as follows:")
for i in range(len(arrivalrates)):
    print(f"For λ = {arrivalrates[i]}, E[N] = {queueslen[i]}, E[T] = {responsetimes[i]}; and E[N] = λ * E[T] = {arrivalrates[i] * responsetimes[i]}.")
print("And so, the Little's Law holds.")
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(arrivalrates, responsetimes, color='lightskyblue', label='E[T]')
ax.plot(arrivalrates, queueslen, color='darkblue', label='E[N]')
ax.set_xlabel('Rate of Arrival (λ)')
ax.set_ylabel('Values')
ax.set_title('M/M/1 Queue Graph')
ax.legend()
plt.show()
```
