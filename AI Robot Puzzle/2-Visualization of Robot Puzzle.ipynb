```python
import networkx as nx
import matplotlib.pyplot as plt
grid = [
    ['X', ' ', 'E'],
    [' ', ' ', ' '],
    ['S', ' ', 'X']
]
G = nx.DiGraph()
rows, cols = len(grid), len(grid[0])
for x in range(rows):
    for y in range(cols):
        if grid[x][y] != 'X':
            G.add_node((x, y))
            if y + 1 < cols and grid[x][y + 1] != 'X':
                G.add_edge((x, y), (x, y + 1))
            if x - 1 >= 0 and grid[x - 1][y] != 'X':
                G.add_edge((x, y), (x - 1, y))
pos = {(x, y): (y, -x) for x, y in G.nodes()}
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20)
nx.draw_networkx_nodes(G, pos, nodelist=[(2, 0)], node_color='green', node_size=3000)
nx.draw_networkx_nodes(G, pos, nodelist=[(0, 2)], node_color='red', node_size=3000)
plt.title("Problem Graph for the Puzzle")
plt.show()
```
